```python
#importando librerías
import pandas as pd

#Cargando la data a un dataframe
url='https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
df=pd.read_json(url)
type(df)
```

```python
df.dtypes
```

```python
#Analizar los tipos de datos de las columnas
df.info()
```

```python
df.shape
```

```python
df
```

```python
#Analizar si existen datos nulos
df.isnull().sum()
```

```python
#Normalizaremos las columnas que traen diccionarios
#Crearemos DataFrames independientes para cada categoría
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])
```

```python
# Concatenamos todo en un solo DataFrame limpio
# Mantenemos 'customerID' y 'Churn' del original
df_final = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)
```

```python
df_final
```

```python
df_final.info()
```

```python
# Convertimos Charges.Total a numérico
# errors='coerce' reemplaza cualquier texto inválido por NaN (nulo)
df_final['Charges.Total'] = pd.to_numeric(df_final['Charges.Total'], errors='coerce')
```

```python
# Verificamos si se crearon nulos durante la conversión
print(f"Nulos encontrados en Charges.Total: {df_final['Charges.Total'].isnull().sum()}")
```

```python
# Si hay nulos, podemos llenarlos con 0 (asumiendo que son clientes nuevos sin cargos totales aún)
df_final['Charges.Total'] = df_final['Charges.Total'].fillna(0)
```

```python
# Verificamos nuevamente si apun hay nulos en Charges.Total
print(f"Nulos encontrados en Charges.Total: {df_final['Charges.Total'].isnull().sum()}")
```

```python
# Comparamos el promedio de cargos mensuales entre los que se quedan y los que se van
analisis_gastos = df_final.groupby('Churn')['Charges.Monthly'].mean()
print(analisis_gastos)
```

```python
df_final['Churn']
```

```python
# Filtramos el DataFrame donde la columna Churn sea exactamente un espacio en blanco
vacios_churn = df_final[df_final['Churn'] == " "]
```

```python
# Mostramos cuántos hay y el listado de los primeros 10
print(f"Total de registros con Churn vacío: {len(vacios_churn)}")
vacios_churn[['customerID', 'tenure', 'Charges.Monthly', 'Churn']].head(10)
```

```python
# Quitamos espacios invisibles a los lados de cada texto
df_final['Churn'] = df_final['Churn'].str.strip()
```

```python
# Reemplazamos los strings que quedaron totalmente vacíos por NaN (nulo real)
import numpy as np
df_final['Churn'] = df_final['Churn'].replace('', np.nan)
```

```python
# Ahora sí, eliminamos los nulos
df_final = df_final.dropna(subset=['Churn']).copy()

# Verificación final: Repetir el groupby
print("--- Grupos actuales en Churn ---")
print(df_final.groupby('Churn')['Charges.Monthly'].mean())
```

```python
print(df_final.isnull().sum())
```

```python
df_final.shape
```

```python
#Cuentas diarias
# Creamos la nueva columna dividiendo el cargo mensual por 30
df_final['Cuentas_Diarias'] = (df_final['Charges.Monthly'] / 30).round(2)
```

```python
# Verificamos los primeros registros para ver la nueva columna
df_final[['customerID', 'Charges.Monthly', 'Cuentas_Diarias']].head()
```

```python
# Comparamos el gasto diario promedio por estado de Churn
print("Gasto diario promedio:")
print(df_final.groupby('Churn')['Cuentas_Diarias'].mean())
```

El Diagnóstico de "Cuentas_Diarias"

Clientes que se quedan (No): Pagan en promedio $2.04 al día.

Clientes que se van (Yes): Pagaban en promedio $2.48 al día.

¿Qué significa esto para Telecom X? Que los clientes que se fugan están pagando, en promedio, un 21.5% más cada día que los que deciden quedarse. Esto refuerza la hipótesis de que el precio es un detonante crítico para la evasión.

```python
# Seleccionamos las columnas numéricas clave
columnas_num = ['tenure', 'Charges.Monthly', 'Charges.Total', 'Cuentas_Diarias']

# Generamos el análisis descriptivo
analisis_stats = df_final[columnas_num].describe()

# Añadimos la mediana explícitamente (que es el percentil 50)
print("Análisis Descriptivo de Clientes:")
display(analisis_stats)
```

```python
import plotly.express as px

# Preparamos los datos contando las categorías
churn_data = df_final['Churn'].value_counts().reset_index()
churn_data.columns = ['Estado', 'Total']

# Creamos el gráfico de dona
fig_pie = px.pie(churn_data,
                 values='Total',
                 names='Estado',
                 hole=0.5,
                 title='<b>Distribución Porcentual de Evasión (Churn)</b>',
                 color='Estado',
                 color_discrete_map={'No': '#2ECC71', 'Yes': '#E74C3C'}) # Verde para permanencia, Rojo para baja

fig_pie.update_traces(textinfo='percent+label', pull=[0, 0.1])
fig_pie.show()
```

```python
# Gráfico de barras para ver cantidades exactas
fig_bar = px.bar(churn_data,
                 x='Estado',
                 y='Total',
                 text='Total',
                 color='Estado',
                 title='<b>Conteo Total de Clientes: Permanencia vs. Baja</b>',
                 labels={'Total': 'Cantidad de Clientes', 'Estado': '¿Hubo Evasión?'},
                 color_discrete_map={'No': '#2ECC71', 'Yes': '#E74C3C'},
                 template="plotly_white")

fig_bar.update_traces(textposition='outside')
fig_bar.show()
```

```python
#Cruce de Variables Categóricas
#Para este análisis,vamos a usar Gráficos de Barras Agrupados. Comparar el "Yes" y el "No" en tres ejes críticos: Género, Método de Pago y Tipo de Contrato.

import plotly.express as px

# Definimos las variables categóricas que queremos explorar
variables_cat = ['gender', 'Contract', 'PaymentMethod', 'InternetService']

for var in variables_cat:
    # Creamos un DataFrame agrupado para cada variable
    df_grouped = df_final.groupby([var, 'Churn']).size().reset_index(name='Cantidad')

    # Generamos el gráfico
    fig = px.bar(df_grouped,
                 x=var,
                 y='Cantidad',
                 color='Churn',
                 barmode='group',
                 title=f'<b>Distribución de Evasión por {var}</b>',
                 color_discrete_map={'No': '#2ECC71', 'Yes': '#E74C3C'},
                 template="plotly_white")
    fig.show()
```

```python
# Para Método de Pago en porcentajes
print("Tasa de Evasión por Método de Pago (%):")
print(pd.crosstab(df_final['PaymentMethod'], df_final['Churn'], normalize='index') * 100)
```

```python
#análisis por antigüedad
import plotly.express as px

# Creamos un histograma interactivo para ver la distribución de meses
fig_tenure = px.histogram(df_final,
                          x="tenure",
                          color="Churn",
                          marginal="box", # Añade un diagrama de caja arriba para ver la mediana
                          title="<b>Distribución de Antigüedad: ¿Cuándo se van los clientes?</b>",
                          labels={'tenure': 'Meses de Antigüedad', 'count': 'Número de Clientes'},
                          color_discrete_map={'No': '#2ECC71', 'Yes': '#E74C3C'},
                          template="plotly_white",
                          barmode='overlay')

fig_tenure.show()
```

# ******Informe Final: Análisis de Evasión (Churn) - Telecom X *********

## 1. Introducción

El objetivo de este análisis es identificar los factores críticos que influyen en ***la evasión de clientes (Churn) de Telecom X***. El Churn representa el porcentaje de suscriptores que cancelan su servicio, lo cual impacta directamente en la rentabilidad de la empresa, ya que el costo de adquisición de un nuevo cliente es significativamente mayor que el de retención.

## 2. Limpieza y Tratamiento de Datos

Para garantizar la calidad del análisis, se realizaron los siguientes pasos:

* **Importación:** Se cargaron los datos desde el archivo JSON proporcionado, el cual contenía información demográfica, de servicios y financiera de 7,043 clientes.
* **Tratamiento de Nulos:** Se identificaron y eliminaron filas con valores vacíos en la columna "Charges.Total", asegurando un dataset final limpio de **7,043 registros**.
* **Ingeniería de Datos:** Se creó la columna **Cuentas_Diarias**, dividiendo el cargo mensual por 30 para obtener una visión más detallada del valor diario de cada cliente.

## 3. Análisis Exploratorio de Datos (EDA)

Durante la exploración, se identificaron patrones clave mediante visualizaciones:

* **Distribución Financiera:** Los clientes que abandonan la compañía pagan, en promedio, **$2.48 diarios**, frente a los **$2.04** de los clientes que permanecen.
* **Impacto del Contrato:** El contrato **"Month-to-month"** presenta el mayor volumen de fugas con **1,655 bajas**, mientras que los contratos anuales muestran una estabilidad casi total.
* **Riesgo por Servicio:** El servicio de **Fibra Óptica** tiene una tasa de evasión alarmante, con casi tantos clientes retirándose como permaneciendo, a diferencia del servicio DSL que es muy estable.
* **Antigüedad:** El riesgo de fuga es crítico en los **primeros 6 meses**. El gráfico de Boxplot muestra que la mediana de antigüedad para los que se van es de apenas **10 meses**.

## 4. Conclusiones e Insights

* **Sensibilidad al Precio:** Los clientes que evaden enfrentan cargos diarios un **21.5% superiores** a los de los clientes retenidos.
* **Falta de Compromiso:** El método de pago **"Electronic check"** tiene la tasa de evasión más alta con un **45.28%**.
* **Neutralidad Demográfica:** El género no es un factor determinante, ya que la distribución de fuga es casi idéntica entre hombres y mujeres.

## 5. Recomendaciones Estratégicas

1. **Fidelización Temprana:** Implementar campañas de "Acompañamiento" durante los primeros 12 meses, que es el periodo de mayor riesgo según el análisis de *tenure*.
2. **Incentivos Contractuales:** Ofrecer descuentos agresivos a los clientes de "Fibra Óptica" que actualmente tienen contrato mes a mes para que migren a contratos de 1 o 2 años.
3. **Automatización de Pagos:** Promover el uso de tarjeta de crédito o transferencia automática, dado que estos métodos tienen tasas de evasión 3 veces menores que el cheque electrónico.
4. **Revisión de Fibra Óptica:** Investigar la calidad técnica o competitividad de precios de la Fibra Óptica, ya que es el servicio con mayor pérdida de usuarios.
